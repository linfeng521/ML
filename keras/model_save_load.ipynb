```python
import tensorflow as tf
import numpy as np
from matplotlib import pyplot as plt
from PIL import Image
from tensorflow import keras
```

```python
air_lake_model_dir = '../models/air_lake.h5'
```

```python
air_lake_model = keras.models.load_model(air_lake_model_dir)
```

```python
air_lake_model.summary()
```

```python
airplane_path = '../models/airplane.jpg'
lake_path = '../models/lake.jpg'
```

```python
airplane = Image.open(airplane_path)
lake = Image.open(lake_path)
```

```python
airplane1 = np.array(airplane)
lake1 = np.array(lake)
```

```python
plt.subplot(121)
plt.imshow(airplane1)
plt.subplot(122)
plt.imshow(lake1)
plt.show()
```

```python
airplane.size
```

```python
airplane.resize((256,256))
```

```python
lake.resize((256,256))
```

```python
airplane2 = np.array(airplane.resize((256,256)))
lake2 = np.array(lake.resize((256,256)))
```

```python
plt.subplot(121)
plt.imshow(airplane2)
plt.subplot(122)
plt.imshow(lake2)
plt.show()
```

```python
airplane3 = np.expand_dims(airplane2,axis=0)
lake3 = np.expand_dims(lake2,axis=0)
```
