```python
import pandas as pd
```

```python
data = pd.read_csv('./Income1.csv')
```

```python
data
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.scatter(data.Education, data.Income,color='red')
plt.title('income',fontsize=24,color='b')
```

```python
x = data.Education
y = data.Income
```

```python
model = tf.keras.Sequential()
```

```python
model.add(tf.keras.layers.Dense(1, input_shape=(1,)))
```

```python
model.summary()   # y = ax + b 初始化两个变量a,b 输入维度1
```

```python
model.compile(optimizer='adam',
              loss='mse'
) # 均方差
```

```python
history = model.fit(x, y, epochs=5000) #开始训练 所有数据训练5000次,需找a,b的最小值
```

```python
model.predict(x)
```

```python
model.predict(pd.Series([20]))
```
