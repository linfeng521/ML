```python
img = cv.imread('images/zixia.jpg',)
img = cv.cvtColor(img,cv.COLOR_BGR2GRAY)
rows,cols = img.shape

M = np.float32([[0.5,0,100],[0,1,100]])
# 仿射变换
dst = cv.warpAffine(img,M,(cols,rows))
M = cv.getRotationMatrix2D((rows/2,cols/2),180,1)
print(M)
dst = cv.warpAffine(img,M,(cols,rows))
plt.imshow(dst)
```

```python
plt.subplot(221),plt.imshow(img2),plt.title("orgin"),plt.axis("off")
plt.subplot(223),plt.imshow(cv.flip(img2,0)),plt.title("x flip"),plt.axis("off") # 0=x 1=y -1=x+y
plt.subplot(222),plt.imshow(cv.flip(img2,1)),plt.title('y flip"'),plt.axis("off")
plt.subplot(224),plt.imshow(cv.flip(img2,-1)),plt.title('x y flip'),plt.axis("off")
```

```python
# 图像模糊
kernel = np.ones((10,10),np.float32)/100
dst = cv.filter2D(img2,-1,kernel)
plt.subplot(121),plt.imshow(img2),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(dst),plt.title('Averaging')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python

import cv2
import numpy as np

img = cv2.imread('me.png',1)
imgInfo = img.shape

height = imgInfo[0]
width = imgInfo[1]

dstHeight = int(height/2)
dstWidth = int(width/2)

dstImage = np.zeros((dstHeight,dstWidth,3),np.uint8)


for i in range(0,dstHeight):
    for j in range(0,dstWidth):
        iNew = int(i*(height*1.0/dstHeight))
        jNew = int(j*(width*1.0/dstWidth))
        dstImage[i,j] = img[iNew,jNew]

cv2.imshow('i love you',dstImage)
cv2.waitKey(0)
```
